casefolding. Fungsi ini memiliki beberapa tujuan dalam pemrosesan teks.

text = text.lower(): Langkah pertama dalam fungsi ini adalah mengubah seluruh teks menjadi huruf kecil (lowercase). Ini dilakukan untuk memastikan konsistensi dalam pemrosesan teks, sehingga kata-kata dengan huruf besar atau kecil dianggap sama.

text = re.sub(r'https?://\S+|www\.\S+', '', text): Ini menggunakan modul re (regular expression) untuk menghapus URL dan tautan web dari teks. Ekspresi reguler https?://\S+ digunakan untuk mencocokkan URL dengan atau tanpa "https", dan www\.\S+ digunakan untuk mencocokkan tautan web yang dimulai dengan "www.".

text = re.sub(r'[-+]?[0-9]+', '', text): Ini menggunakan ekspresi reguler untuk menghapus angka dari teks. Ekspresi reguler [-+]?[0-9]+ mencocokkan angka positif atau negatif dalam teks dan menghapusnya.

text = re.sub(r'[^\w\s]', '', text): Ini menggunakan ekspresi reguler [^\w\s] untuk menghapus karakter selain huruf dan spasi dari teks. Dengan kata lain, ini menghapus semua karakter khusus seperti tanda baca dan simbol.

text = text.strip(): Terakhir, fungsi ini menghapus spasi tambahan dari awal dan akhir teks (jika ada).

Hasil akhir dari fungsi casefolding adalah teks yang telah melalui serangkaian transformasi untuk mengubahnya menjadi format yang lebih bersih dan konsisten, dengan menghilangkan URL, angka, karakter khusus, dan spasi tambahan, serta mengonversinya menjadi huruf kecil (lowercase). Fungsi ini berguna dalam pemrosesan teks untuk analisis atau pemodelan data.

```python
import re

def punctuation(text):
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'[-+]?[0-9]+', '',text)
    text = re.sub(r'[^\w\s]', '',text)
    text = text.strip()
    return text
```

more_stopword = ['tsel','gb','rb','btw']: Ini adalah daftar kata-kata tambahan yang ingin Anda tambahkan ke daftar kata-kata stop bahasa Indonesia yang telah diunduh sebelumnya. Misalnya, Anda ingin menjadikan kata-kata seperti "tsel," "gb," "rb," dan "btw" sebagai kata-kata stop tambahan yang akan dihapus dari teks.

stopwords_ind = stopwords_ind + more_stopword: Dalam langkah ini, Anda menggabungkan daftar kata-kata stop tambahan more_stopword ke dalam daftar stopwords_ind yang sudah ada. Ini memperluas daftar kata-kata stop untuk mencakup kata-kata yang Anda tambahkan.

def remove_stop_word(text): Ini adalah definisi fungsi remove_stop_word yang akan digunakan untuk menghapus kata-kata stop dari teks yang diberikan.

clean_words = []: Ini adalah list kosong yang akan digunakan untuk menyimpan kata-kata yang tidak termasuk dalam daftar kata-kata stop.

text = text.split(): Teks input dipecah menjadi kata-kata dengan menggunakan spasi sebagai pemisah.

for word in text:: Fungsi ini akan melakukan perulangan melalui setiap kata dalam teks.

if word not in stopwords_ind:: Setiap kata akan diperiksa apakah ada dalam daftar kata-kata stop (stopwords_ind). Jika kata tersebut tidak ada dalam daftar, maka kata tersebut akan dimasukkan ke dalam list clean_words.

return " ".join(clean_words): Fungsi ini menggabungkan kata-kata yang tersimpan dalam list clean_words menjadi satu string dengan menggunakan spasi sebagai pemisah, dan string tersebut akan dikembalikan sebagai output.

Dengan menggunakan fungsi remove_stop_word ini, Anda dapat menghapus kata-kata stop dan kata-kata tambahan yang Anda tentukan dari teks input Anda sehingga mendapatkan teks yang lebih bersih dan relevan untuk analisis lanjutan.

```python
# Membuat fungsi stop removal

#Menambahkan kata dalam stopword

more_stopword = ['tsel','gb','rb','btw']
stopwords_ind = stopwords_ind + more_stopword

def remove_stop_word(text):
    clean_words =[]
    text = text.split()
    for word in text:
        if word not in stopwords_ind:
            clean_words.append(word)
    return " ".join(clean_words)
```

```python
# Define a function for text preprocessing
def text_preprocessing_process(text):
    # Remove punctuation
    text = punctuation(text)
   
    # Lowercase the text
    text = casefolding(text)

    #Text Normalization
    text = text_normalize(text)

    text = stemming(text)

    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words('indonesian'))
    stop_words.remove("tidak")
    stop_words.remove("enggak")
    tokens = [word for word in tokens if word not in stop_words]

    # Lemmatize the tokens
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    # Join the tokens back into a single string
    text = ' '.join(tokens)

    return text
```
