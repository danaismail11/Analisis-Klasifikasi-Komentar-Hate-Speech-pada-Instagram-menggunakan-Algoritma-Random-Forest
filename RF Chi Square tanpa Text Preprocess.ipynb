from sklearn.feature_extraction.text import TfidfVectorizer: Ini mengimpor modul TfidfVectorizer dari library scikit-learn (sklearn). TfidfVectorizer digunakan untuk mengubah teks menjadi vektor TF-IDF, yang digunakan dalam pemrosesan teks dan analisis teks.

vec_TF_IDF = TfidfVectorizer(ngram_range=(1,1)): Anda membuat sebuah objek vec_TF_IDF yang merupakan instance dari TfidfVectorizer. Parameter ngram_range=(1,1) menunjukkan bahwa Anda ingin menggunakan model unigram, yang berarti hanya kata-kata tunggal yang akan dihitung frekuensinya.

vec_TF_IDF.fit(x): Anda menggunakan objek vec_TF_IDF untuk melakukan fitting pada data x. Ini akan menghitung nilai TF-IDF untuk setiap kata dalam data x.

x_tf_idf = vec_TF_IDF.transform(x): Anda menggunakan objek vec_TF_IDF yang telah di-fit untuk mengubah data x menjadi vektor TF-IDF.

pickle.dump(vec_TF_IDF.vocabulary_, open("feature_tf-idf.sav", "wb")): Dalam langkah ini, Anda menggunakan pickle.dump untuk menyimpan kamus kata-kata yang terkait dengan vektor TF-IDF ke dalam file "feature_tf-idf.sav" dalam mode binary write ("wb"). Kamus ini akan berguna saat Anda ingin mengonversi teks baru menjadi vektor TF-IDF dengan kata-kata yang sama yang telah digunakan dalam training.

Dengan kode ini, Anda telah menyimpan model TF-IDF dan kamus kata-kata yang sesuai ke dalam file "feature_tf-idf.sav", yang dapat Anda gunakan untuk transformasi teks baru ke dalam vektor TF-IDF menggunakan kamus yang sama.

```python
# Save Model
import pickle

#TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

#Unigram
vec_TF_IDF = TfidfVectorizer(ngram_range=(1,2))
vec_TF_IDF.fit(x)

x_tf_idf =vec_TF_IDF.transform(x)

pickle.dump(vec_TF_IDF.vocabulary_,open("feature_tf-idf.sav", "wb"))
```

Kode yang Anda berikan adalah contoh penggunaan library scikit-learn (sklearn) untuk melakukan seleksi fitur dengan metode chi-squared (chi2). Ini berguna untuk mengurangi jumlah fitur dalam data Anda menjadi fitur-fitur yang paling relevan. Mari saya jelaskan langkah-langkahnya:

from sklearn.feature_selection import SelectKBest: Ini mengimpor kelas SelectKBest dari sklearn. SelectKBest adalah algoritma seleksi fitur yang memungkinkan Anda untuk memilih fitur-fitur terbaik berdasarkan metrik tertentu.

from sklearn.feature_selection import chi2: Ini mengimpor metrik chi-squared (chi2) yang akan digunakan untuk menilai relevansi fitur.

chi2_features = SelectKBest(chi2, k=2000): Anda membuat sebuah objek chi2_features dari kelas SelectKBest dan mengaturnya untuk menggunakan metrik chi-squared (chi2) dengan memilih 2000 fitur terbaik (k=2000). Ini berarti Anda ingin mempertahankan 2000 fitur terbaik berdasarkan perhitungan chi-squared.

x_kbest_features = chi2_features.fit_transform(x_train, y_train): Anda menggunakan objek chi2_features yang telah dibuat untuk melakukan seleksi fitur pada data pelatihan (x_train dan y_train). Hasil dari seleksi ini disimpan dalam variabel x_kbest_features, yang akan berisi data dengan hanya 2000 fitur terbaik yang telah dipilih.

print("Original Feature Number : ", x_train.shape[1]): Ini mencetak jumlah fitur asli dalam data pelatihan sebelum seleksi fitur dilakukan.

print("Reduced Feature Number: ", x_kbest_features.shape[1]): Ini mencetak jumlah fitur setelah seleksi fitur dilakukan, yang dalam hal ini adalah 2000 fitur terbaik yang dipilih oleh metode chi-squared.

Dengan menggunakan SelectKBest dengan metode chi-squared, Anda dapat mengurangi dimensi data Anda dan mempertahankan hanya fitur-fitur yang paling informatif untuk tugas analisis atau pemodelan berikutnya.

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

chi2_features =SelectKBest(chi2, k=276)
x_kbest_features =chi2_features.fit_transform(x_train, y_train)

print("Original Feature Number : ", x_train.shape[1])
print("Reduced Feature Number: ",x_kbest_features.shape[1])
```
